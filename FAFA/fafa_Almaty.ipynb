### 1. Which category brings the most revenue?

1) Find Revenue for all Categories and make a bar plot to compare.

```python
# Iterate through an array of product rays and find the sum of prices, which is a Revenue for all 9 Categories.

Revenues = {}

Revenues ={str(i) : data[data['Product Name'] == str(i)]['Price(TNG)'].sum() for i in array_ProductNames}

print(Revenues)
```

```python
names = list(Revenues.keys())
values = list(Revenues.values())

sns_t = sns.barplot(x=values, y = names)
plt.xlabel("Revenue")
plt.ylabel("Category")
plt.title('Revenues for all Categories')

# Function to add Number Labels on Each Bar

def show_values_on_bars(axs, h_v="v", space=0.4):
    def _show_on_single_plot(ax):
        if h_v == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height()
                value = int(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif h_v == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height()
                value = int(p.get_width())
                ax.text(_x, _y, value, ha="left", va="center", fontsize = 7)

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _show_on_single_plot(ax)
    else:
        _show_on_single_plot(axs)

show_values_on_bars(sns_t , "h", 0.3)

plt.show()
```

## ANSWER: Building Materials and TNP give the most revenue. 
#### 1. Building - 9 035 160 TNG,
#### 2. TNP - 8 570 150 TNG.

Other Categories yield a lot less Revenue.

#### 2. What trucks are most needed for the most popular Categories?

```python
# 1) BUILDING CATEGORY

Trucks_Building = data[data['Product Name'] == 'Стройматериалы']
Trucks_Building_bar = Trucks_Building['Special Request'].value_counts().plot(kind='barh')
show_values_on_bars(Trucks_Building_bar, "h", 0.2)
```

#### ANSWER: Building Category mostly needs tent trucks, and if we dont want to know which parts of the trucks are needed, we can simplify.

```python
Trucks_Building_bar = Trucks_Building['Special Request'].str[:5].value_counts().plot(kind='barh')
show_values_on_bars(Trucks_Building_bar, "h", 0.2)
```

#### > 37 of the time we need TENT Trucks, others are a lot less.

```python
# 2) TNT CATEGORY

Trucks_TNP = data[data['Product Name'] == 'ТНП']
Trucks_TNP_pie = Trucks_TNP['Special Request'].value_counts()
# show_values_on_bars(Trucks_TNP_bar, "h", 0.2)

plt.pie(Trucks_TNP_pie, labels = Trucks_TNP_pie.keys())
plt.title('Trucks for TNT')
# plt.legend()
plt.show() 
```

```python
Trucks_TNP = data[data['Product Name'] == 'ТНП']
Trucks_TNP_bar = Trucks_TNP['Special Request'].str[:5].value_counts().plot(kind='barh')
show_values_on_bars(Trucks_TNP_bar, "h", 0.2)
```

## 3. What are the popular destinations for the most popular Categories?

```python
# 1) BUILDING CATEGORY

Trucks_Building = data[data['Product Name'] == 'Стройматериалы']
Trucks_Building_bar = Trucks_Building['Destination'].str.partition('→')[2].str.partition('-')[0].value_counts().plot(kind='barh')
show_values_on_bars(Trucks_Building_bar, "h", 0.2)
plt.title('Popular Destinations for Building')
```

### ANSWER: SHYMKENT is and NurSultan are the most popular for Building Category.

```python
# 2) TNP CATEGORY

Trucks_TNP = data[data['Product Name'] == 'ТНП']
Trucks_TNP_bar = Trucks_TNP['Destination'].str.partition('→')[2].str.partition('-')[0].value_counts().plot(kind='barh')
show_values_on_bars(Trucks_TNP_bar, "h", 0.2)
plt.title('Popular Destinations for TNP')
```

### ANSWER: SHYMKENT is the most popular for the TNP Category.
